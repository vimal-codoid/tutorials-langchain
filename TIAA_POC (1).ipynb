```python
import pandas as pd
main_sheet = pd.read_csv('Table1.csv')
mapping_sheet = pd.read_csv('Table2.csv')
```

```python
main_sheet
```

```python
merged_df = pd.merge(main_sheet, mapping_sheet, left_on='Account', right_on='Assets', how='inner')
```

```python
filtered_df = merged_df[['Account','Assets Category']]
```

```python
lst1 = filtered_df['Account'].to_list()
lst2 = filtered_df['Assets Category'].to_list()
```

```python
my_dict = dict(zip(lst1, lst2))
```

```python
main_sheet['Account'] = main_sheet['Account'].replace(my_dict)
```

```python
main_sheet = main_sheet[main_sheet['Account'] != 'Sub Total:']
```

```python
# Remove commas from the 'ColumnName' values
main_sheet['Opening Balance'] = main_sheet['Opening Balance'].str.replace(',', '')

# Convert the 'ColumnName' values to integers
main_sheet['Opening Balance'] = main_sheet['Opening Balance'].astype(float)
```

```python
grouped = main_sheet.groupby('Account')
```

```python
result = grouped.sum()
```

```python
result = result.reset_index()
```

```python
result
```

```python
result['Opening Balance'].sum()
```

```python
new_row = {'Account': 'Total', 'Opening Balance': 13926330.53 }

# Append the new row to the DataFrame
result = result.append(new_row, ignore_index=True)
```

```python
import pandas as pd
column_names = ['main', 'map']

df = pd.read_csv('/content/Table2_combined_1.csv', header = None, names = column_names)
```

```python
df
```

```python
# df = pd.DataFrame(data)

# Find the indices where 'Assets' and 'Liabilities' occur
assets_index = df[df['main'] == 'Assets Category'].index[0]
liabilities_index = df[df['main'] == 'Liabilities Category'].index[0]

# Split the DataFrame based on the indices
df_assets = df.iloc[assets_index + 1:liabilities_index]
df_liabilities = df.iloc[liabilities_index + 1:]

# Reset the index for both DataFrames
df_assets.reset_index(drop=True, inplace=True)
df_liabilities.reset_index(drop=True, inplace=True)

# Print the result
print("DataFrame for Assets:")
print(df_assets)
```
